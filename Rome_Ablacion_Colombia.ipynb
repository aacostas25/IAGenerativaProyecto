#4. Cargar LLaMA-3 8B Instruct con EasyEdit para ROME

```python
# ==========================================
# 4) CONFIGURAR LLaMA-3 8B PARA EASYEDIT + ROME
# ==========================================

LLAMA3_MODEL_NAME = "meta-llama/Meta-Llama-3-8B-Instruct"

# Usamos el archivo hparams EXACTO que sí existe en tu instalación
# (lo verificaste en tu lista de ./hparams/ROME/)
hparams_llama3 = ROMEHyperParams.from_hparams("./hparams/ROME/llama3-8b")

# Sobrescribimos el nombre del modelo para usar LLaMA-3 de HuggingFace
hparams_llama3.model_name = LLAMA3_MODEL_NAME

# Si el objeto tiene tokenizer_name lo alineamos
if hasattr(hparams_llama3, "tokenizer_name"):
    hparams_llama3.tokenizer_name = LLAMA3_MODEL_NAME

# ==========================================
# CORRECCIÓN DE ERROR DEL DISPOSITIVO
# (El error 'cuda:cuda' se debe a string incorrecto)
# ==========================================
if torch.cuda.is_available():
    hparams_llama3.device = 0       # EasyEdit construye "cuda:0"
else:
    hparams_llama3.device = "cpu"   # CPU si no hay GPU

# ==========================================
# Carga en 4-bit para que LLaMA-3 8B quepa en Colab
# ==========================================
hparams_llama3.load_in_4bit = True
hparams_llama3.load_in_8bit = False

# ==========================================
# Crear el editor (modelo + tokenizer + configs)
# ==========================================
editor_llama = BaseEditor.from_hparams(hparams_llama3)

print("LLaMA-3 8B cargado correctamente en EasyEdit para edición con ROME.")


```

#5. Tokenizer y función para generar texto

```python
# ==========================================
# 5) FUNCIÓN PARA CONSULTAR A LLaMA-3 8B
# ==========================================
from transformers import AutoTokenizer

tokenizer_llama = AutoTokenizer.from_pretrained(LLAMA3_MODEL_NAME)

def generar_llama(prompt, max_new_tokens=300, temperature=0.001, top_p=1):
    """
    Genera texto con LLaMA-3 8B Instruct en formato chat,
    evitando que repita el prompt.
    """
    device = editor_llama.model.device

    # 1) Armamos el historial de chat
    messages = [
        {
            "role": "system",
            "content": (
                "Eres un narrador experto en mitos y leyendas de América Latina. "
                "Respondes en español, con explicaciones claras y coherentes."
                "Responde de manera breve y concisa solamente sobre lo que se te esta preguntando"
            ),
        },
        {
            "role": "user",
            "content": prompt,
        },
    ]

    # 2) Convertimos a texto usando el chat_template de LLaMA-3
    chat_input = tokenizer_llama.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    # 3) Tokenizamos y enviamos al modelo
    inputs = tokenizer_llama(chat_input, return_tensors="pt").to(device)

    output_ids = editor_llama.model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        temperature=temperature,
        top_p=top_p,
        repetition_penalty=1.1,  # ayuda a evitar loops
        eos_token_id=tokenizer_llama.eos_token_id,
        pad_token_id=tokenizer_llama.eos_token_id,
    )

    # 4) Nos quedamos solo con la parte generada (sin el prompt)
    generated_ids = output_ids[0][inputs["input_ids"].shape[1]:]
    text = tokenizer_llama.decode(generated_ids, skip_special_tokens=True)

    return text.strip()
```

#7. Preparar los requests para ROME desde el JSON

```python
# ==========================================
# 7) CONSTRUIR REQUESTS PARA ROME
# ==========================================

def construir_requests_rome(mitos):
    requests = []
    for m in mitos:
        req = {
            "prompt": m["prompt"],
            "subject": m["subject"],
            "target_new": m["target_new"],
        }
        if "target_true" in m:
            req["target_true"] = m["target_true"]
        requests.append(req)
    return requests

requests_chile = construir_requests_rome(mitos_chile)
print("Ejemplo de request para ROME:", requests_chile[0])
print("Total requests:", len(requests_chile))
```

##0) Bloque común (helpers) – ejecútalo una vez por sesión

```python
# ==========================================
# BLOQUE COMÚN: HELPERS PARA LOS EXPERIMENTOS
# ==========================================
import copy
import gc
import torch
import pandas as pd
from easyeditor import BaseEditor, ROMEHyperParams

def clonar_hparams(hparams_base):
    """
    Crea una copia profunda de los hparams base,
    para no modificar el objeto original.
    """
    return copy.deepcopy(hparams_base)

def obtener_capa_base(hparams):
    """
    Intenta leer la capa base desde los hparams para documentar el experimento.
    """
    if hasattr(hparams, "layers") and isinstance(hparams.layers, (list, tuple)) and len(hparams.layers) > 0:
        return hparams.layers[0]
    if hasattr(hparams, "layer"):
        return hparams.layer
    return None

def setear_capa_en_hparams(hparams, layer_idx):
    """
    Ajusta la capa donde ROME aplicará la edición.
    Trata de ser robusto a diferentes nombres de atributo.
    """
    if layer_idx is None:
        return hparams

    if hasattr(hparams, "layers") and isinstance(hparams.layers, (list, tuple)):
        hparams.layers = [layer_idx]
        print(f"[INFO] hparams.layers fijado a {hparams.layers}")
        return hparams

    if hasattr(hparams, "layer"):
        hparams.layer = layer_idx
        print(f"[INFO] hparams.layer fijado a {hparams.layer}")
        return hparams

    print("[ADVERTENCIA] No se encontró atributo de capa ('layers' o 'layer') en hparams.")
    return hparams

def correr_experimento(nombre_exp, hparams_base, rome_request_exp):
    """
    Crea un editor a partir de hparams_base,
    corre una edición con ROME y extrae métricas
    de reliability (rewrite_acc) y locality.
    """
    print(f"\n==============================")
    print(f"Ejecutando experimento: {nombre_exp}")
    print(f"==============================")

    # 1) Crear editor desde hparams (modelo + tokenizer + configs)
    editor = BaseEditor.from_hparams(hparams_base)

    # 2) Ejecutar ROME sobre el request indicado
    metrics = editor.edit_requests(
        requests=[rome_request_exp],
        sequential_edit=False,
        verbose=True
    )

    # 3) Normalizar formato de salida
    if isinstance(metrics, tuple):
        all_metrics = metrics[0]
    else:
        all_metrics = metrics

    if isinstance(all_metrics, list):
        m = all_metrics[0]
    else:
        m = all_metrics

    pre = m.get("pre", {})
    post = m.get("post", {})

    rewrite_acc_pre = pre.get("rewrite_acc", None)
    rewrite_acc_post = post.get("rewrite_acc", None)

    locality_pre = pre.get("locality", None)
    locality_post = post.get("locality", None)

    # 4) Liberar modelo y memoria
    del editor
    torch.cuda.empty_cache()
    gc.collect()

    resultados = {
        "experimento": nombre_exp,
        "rewrite_acc_pre": rewrite_acc_pre,
        "rewrite_acc_post": rewrite_acc_post,
        "locality_pre": locality_pre,
        "locality_post": locality_post,
    }

    print("Resultados experimento:", resultados)
    return resultados
```

## Script 1 – Experimento baseline (ROME original)

```python
# ==========================================
# EXPERIMENTO 1: BASELINE ROME
# ==========================================

resultados_exp1 = []

capa_base = obtener_capa_base(hparams_llama3)
print(f"Capa base de ROME (según hparams_llama3): {capa_base}")

hparams_baseline = clonar_hparams(hparams_llama3)

resultado_base = correr_experimento(
    nombre_exp=f"baseline_rome_capa_{capa_base}",
    hparams_base=hparams_baseline,
    rome_request_exp=rome_request
)

resultados_exp1.append(resultado_base)

df_exp1 = pd.DataFrame(resultados_exp1)
print("\nRESULTADOS EXPERIMENTO 1 (BASELINE):")
print(df_exp1)

# Si quieres, guarda a CSV
df_exp1.to_csv("exp1_baseline_rome.csv", index=False)
```

## Script 2 – Experimento “sin locality” (locality vacío)

```python
# ==========================================
# EXPERIMENTO 2: SIN LOCALITY (LOCALITY DUMMY)
# ==========================================

resultados_exp2 = []

capa_base = obtener_capa_base(hparams_llama3)
print(f"Capa base de ROME (según hparams_llama3): {capa_base}")

# Clonamos el request original y definimos locality dummy
rome_request_sin_locality = copy.deepcopy(rome_request)

# IMPORTANTE: EasyEdit NO acepta listas vacías. Usamos un dummy mínimo.
rome_request_sin_locality["locality"] = {
    "sentences": {
        "prompt": ["Esta frase no tiene relación con el mito."],
        "ground_truth": ["irrelevante"]
    }
}

hparams_sin_locality = clonar_hparams(hparams_llama3)

resultado_sin_locality = correr_experimento(
    nombre_exp=f"sin_locality_capa_{capa_base}",
    hparams_base=hparams_sin_locality,
    rome_request_exp=rome_request_sin_locality
)

resultados_exp2.append(resultado_sin_locality)

df_exp2 = pd.DataFrame(resultados_exp2)
print("\nRESULTADOS EXPERIMENTO 2 (SIN LOCALITY):")
print(df_exp2)

df_exp2.to_csv("exp2_sin_locality.csv", index=False)

```
